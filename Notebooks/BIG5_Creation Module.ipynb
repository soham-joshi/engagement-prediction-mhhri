### BIG 5 Traits

#### Acquaintance

```python
import pandas as pd
```

```python
Acquaintances_Scores={"U00"+ str(i) if(i<10) else "U0"+ str(i) : [0 for _ in range(5)] for i in range(1,19)}
```

```python
for person in Acquaintances_Scores.keys():
    data=pd.read_csv(person + ".csv",header=None)
    data["agg"]=data.sum(axis=1)
    Ac_count=len(data.columns)-1
    data.loc[[0,2,3,4,6]]["agg"]=(10*Ac_count) -data.loc[[0,2,3,4,6]]["agg"]
    Acquaintances_Scores[person][0]=(data["agg"][0]+data["agg"][5])/(2*Ac_count)
    Acquaintances_Scores[person][1]=(data["agg"][1]+data["agg"][6])/(2*Ac_count)
    Acquaintances_Scores[person][2]=(data["agg"][2]+data["agg"][7])/(2*Ac_count)
    Acquaintances_Scores[person][3]=(data["agg"][3]+data["agg"][8])/(2*Ac_count)
    Acquaintances_Scores[person][4]=(data["agg"][4]+data["agg"][9])/(2*Ac_count)    
```

```python
Acquaintances_Scores
```

```python
aggregation_list = []
for user, user_score in Acquaintances_Scores.items():
    its_score = [user]
    its_score.extend(user_score)
    aggregation_list.append(its_score)
    
Acquaintance_traits_DF = pd.DataFrame(aggregation_list, columns = ['user', 'Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness'])
Acquaintance_traits_DF.head()
    
```

```python
Acquaintance_traits_DF.to_csv('Acquaintance_Assessment_Big5_Traits.csv', index = False)
```

```python
Self_DF = pd.read_csv("./personality_labels.txt",header=None,delimiter=" ")
Self_DF.head()
```

```python
Sl_Scores={i:[0 for _ in range(5)]for i in range(1,19)}
```

```python
Self_DF=Self_DF.drop(columns=[0])
Self_DF.head()
```

```python
Self_DF=Self_DF.astype(int)
```

```python
for person in Sl_Scores.keys():
    Sl_Scores[person][0]=(10-Self_DF[1][person-1]+Self_DF[6][person-1])/(2)
    Sl_Scores[person][1]=(10+Self_DF[2][person-1]-Self_DF[7][person-1])/(2)
    Sl_Scores[person][2]=(10-Self_DF[3][person-1]+Self_DF[8][person-1])/(2)
    Sl_Scores[person][3]=(10-Self_DF[4][person-1]+Self_DF[9][person-1])/(2)
    Sl_Scores[person][4]=(10-Self_DF[5][person-1]+Self_DF[10][person-1])/(2)
```

```python
Sl_Scores
```

```python
Self_Scores = []
for user, user_score in Sl_Scores.items():
    user = "U00" + str(user) if user < 10 else "U0" + str(user)
    its_score = [user]
    its_score.extend(user_score)
    Self_Scores.append(its_score)
```

```python
Self_Scores
```

```python
Self_traits_DF = pd.DataFrame(Self_Scores, columns = ['user', 'Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness'])
Self_traits_DF.head()
```

```python
Self_traits_DF.to_csv('Self_Assessment_Big5_Traits.csv', index = False)
```

```python
Self_traits_DF = pd.read_csv('Self_Assessment_Big5_Traits.csv')
Self_traits_DF.head()
```

### Create Big 5 Traits - weighted mean of self and acquaintance

```python
Big_5_aggregated = pd.DataFrame(columns = Self_traits_DF.columns)
```

```python
list_b5_t = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']
Big_5_aggregated['user'] = Self_traits_DF['user']
alpha = 0.8
Big_5_aggregated[list_b5_t] =  alpha * Self_traits_DF[list_b5_t] + (1.0 - alpha) * Acquaintance_traits_DF[list_b5_t]
Big_5_aggregated.head()
```

```python
Big_5_aggregated.to_csv('../Big_5_agg.csv', index = False)
```

### Categorizing

```python
list_b5_t = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']
Big_5_weighted_categorised = Big_5_aggregated
Big_5_weighted_categorised[list_b5_t] = Big_5_aggregated[list_b5_t].round(0).astype(int)
Big_5_weighted_categorised.head()
```

```python
Big_5_weighted_categorised.to_csv('Big_5_weighted_categorised.csv', index = False)
```
