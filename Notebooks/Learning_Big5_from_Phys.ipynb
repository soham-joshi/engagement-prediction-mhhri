## Learning Big 5 Traits from Physiological Features

#### For HHI setting

### Self DF

```python
import pandas as pd
import numpy as np
```

```python
!pip install sklearn
```

```python
HHI_Phy_DF = pd.read_csv('Affectiva/HHI_Affectiva_Recordings/HHI_Physiological_Feature_Vectors.csv')
print(HHI_Phy_DF.shape)
HHI_Phy_DF.head()
```

```python
list_phy_cols = ['phy_feature_' + str(i) for i in range(1, 31)]
list_b5_t = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']
```

```python
Big_5_weighted = pd.read_csv('Self_Assessment_Personality/Self_Assessment_Big5_Traits.csv')
Big_5_weighted.tail()
```

```python
Big_5_weighted_categorised = Big_5_weighted
Big_5_weighted_categorised.tail()
```

```python
Big_5_weighted_binarised = Big_5_weighted_categorised
Big_5_weighted_binarised[list_b5_t] = (Big_5_weighted_categorised[list_b5_t] >= Big_5_weighted_categorised[list_b5_t].mean()).astype(int)
Big_5_weighted_binarised.head()
```

```python
# Big_5_weighted_binarised.to_csv('Big_5_weighted_binarised.csv', index = False)
```

```python
# Big_5_weighted_binarised = pd.read_csv('Big_5_weighted_binarised.csv')
```

```python
Big_5_weighted_binarised.tail()
```

### Creating a Dataset

```python
HHI_Phy_Big5_combined = HHI_Phy_DF.merge(Big_5_weighted_binarised, left_on = 'user', right_on = 'user')
HHI_Phy_Big5_combined.tail(60)
```

```python
HHI_Phy_Big5_combined.tail()
```

```python
HHI_Phy_Big5_combined.shape
```

```python
# HHI_Phy_Big5_combined_shuffled = HHI_Phy_Big5_combined.sample(frac = 1)
# HHI_Phy_Big5_combined_shuffled.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
HHI_Phy_Big5_final = HHI_Phy_Big5_combined
HHI_Phy_Big5_final[list_phy_cols] = scaler.fit_transform(HHI_Phy_Big5_combined[list_phy_cols])
```

```python
HHI_Phy_Big5_final.head(5)
```

### Extraversion

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

```python
X = HHI_Phy_Big5_final[list_phy_cols].to_numpy()
Y = HHI_Phy_Big5_final['Extraversion'].to_numpy()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.13, shuffle=False)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(X_train, Y_train)
```

```python
clf.score(X_test, Y_test)
```

```python
Y_preds = clf.predict(X_test)
Y_preds
```

```python
print(classification_report(Y_test, Y_preds))
```

```python
scores = cross_val_score(clf, X, Y, cv=18)
print(scores)
print(np.mean(scores))
```
