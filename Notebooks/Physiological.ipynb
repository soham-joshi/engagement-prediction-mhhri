```python
def produce_feature_vector(filename):
    
    feature_vector = [0]*30
    
    subject_clip = sio.loadmat(filename)
    subject_z = subject_clip['data'][0][0][1]
    subject_y = subject_clip['data'][0][0][2]
    subject_x = subject_clip['data'][0][0][3]
    subject_cel = subject_clip['data'][0][0][4]
    subject_eda = subject_clip['data'][0][0][5]
    
    # Z
    subject_z_min = np.min(subject_z)
    subject_z_max = np.max(subject_z)
    subject_z_mean = np.mean(subject_z)
    subject_z_std = np.std(subject_z)
    subject_z_grad1 = np.gradient(subject_z, axis = 0)
    subject_z_grad1_mean = np.mean(subject_z_grad1)
    subject_z_grad2 = np.gradient(subject_z_grad1, axis = 0)
    subject_z_grad2_mean = np.mean(subject_z_grad2)
    
    feature_vector[0:6] = [subject_z_min, subject_z_max, subject_z_mean, subject_z_std, subject_z_grad1_mean, subject_z_grad2_mean]
    
    # Y
    subject_y_min = np.min(subject_y)
    subject_y_max = np.max(subject_y)
    subject_y_mean = np.mean(subject_y)
    subject_y_std = np.std(subject_y)
    subject_y_grad1 = np.gradient(subject_y, axis = 0)
    subject_y_grad1_mean = np.mean(subject_y_grad1)
    subject_y_grad2 = np.gradient(subject_y_grad1, axis = 0)
    subject_y_grad2_mean = np.mean(subject_y_grad2)
    
    feature_vector[6:12] = [subject_y_min, subject_y_max, subject_y_mean, subject_y_std, subject_y_grad1_mean, subject_y_grad2_mean]
    
    # X
    subject_x_min = np.min(subject_x)
    subject_x_max = np.max(subject_x)
    subject_x_mean = np.mean(subject_x)
    subject_x_std = np.std(subject_x)
    subject_x_grad1 = np.gradient(subject_x, axis = 0)
    subject_x_grad1_mean = np.mean(subject_x_grad1)
    subject_x_grad2 = np.gradient(subject_x_grad1, axis = 0)
    subject_x_grad2_mean = np.mean(subject_x_grad2)
    
    feature_vector[12:18] = [subject_x_min, subject_x_max, subject_x_mean, subject_x_std, subject_x_grad1_mean, subject_x_grad2_mean]
    
    # cel
    subject_cel_min = np.min(subject_cel)
    subject_cel_max = np.max(subject_cel)
    subject_cel_mean = np.mean(subject_cel)
    subject_cel_std = np.std(subject_cel)
    subject_cel_grad1 = np.gradient(subject_cel, axis = 0)
    subject_cel_grad1_mean = np.mean(subject_cel_grad1)
    subject_cel_grad2 = np.gradient(subject_cel_grad1, axis = 0)
    subject_cel_grad2_mean = np.mean(subject_cel_grad2)
    
    feature_vector[18:24] = [subject_cel_min, subject_cel_max, subject_cel_mean, subject_cel_std, subject_cel_grad1_mean, subject_cel_grad2_mean]
    
    # eda
    subject_eda_min = np.min(subject_eda)
    subject_eda_max = np.max(subject_eda)
    subject_eda_mean = np.mean(subject_eda)
    subject_eda_std = np.std(subject_eda)
    subject_eda_grad1 = np.gradient(subject_eda, axis = 0)
    subject_eda_grad1_mean = np.mean(subject_eda_grad1)
    subject_eda_grad2 = np.gradient(subject_eda_grad1, axis = 0)
    subject_eda_grad2_mean = np.mean(subject_eda_grad2)
    
    feature_vector[24:] = [subject_eda_min, subject_eda_max, subject_eda_mean, subject_eda_std, subject_eda_grad1_mean, subject_eda_grad2_mean]
    
    return feature_vector
    
```

```python
%cd Affectiva/HHI_Affectiva_Recordings
```

```python
directory = '.'
clip_count = 0
# HHI_Phy_DF = pd.DataFrame(columns= ['user'] + ['phy_feature_' + str(i) for i in range(1, 30)])
# # HHI_Phy_DF=HHI_Phy_DF.astype(object)
features_collection = []


for filename in os.listdir(directory):
    if filename.endswith(".mat"): 
#         print(os.path.join(directory, filename))
        parsed_filename = filename.split('_')
        session = parsed_filename[0]
        user_id = parsed_filename[1]
        user_fv = produce_feature_vector(filename)
        its_feature_list = [user_id] + user_fv
        features_collection.append(its_feature_list)
#         HHI_Phy_DF.at[clip_count, 'phy_feature_vector']= list(user_fv)
#         HHI_Phy_DF.at[clip_count, 'user']= user_id
        clip_count = clip_count + 1

HHI_Phy_DF = pd.DataFrame(features_collection, columns= ['user'] + ['phy_feature_' + str(i) for i in range(1, 31)])
```
